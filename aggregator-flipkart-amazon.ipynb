```python
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd
from time import sleep
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
import socket

driver = webdriver.Chrome("/Users/Anil/chromedriver")

prices1=[]
prices2=[]
names1=[]
names2=[]
c=[]
links1=[]
links2=[]


name = input("enter product name \n")
sites = ['flipkart','amazon']
fin=[]
comp=[]
comp2=[]
comp3=[]
cntr=0
nc = name.strip().replace(' ','+')


webs=["https://www.amazon.in/s?k=NAME&ref=nb_sb_noss","https://www.flipkart.com/search?q=NAME&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off"]
for i in webs:   
    fin.append(i.replace("NAME",nc))
    
def amzp():
    driver.get(fin[0])

    content = driver.page_source
    soup = BeautifulSoup(content)
    
    
    
    for div in soup.findAll('a',href=True, attrs={'class':'a-link-normal a-text-normal'}):
        name1 = div.find('span')#, attrs={'class':'a-size-medium a-color-base a-text-normal'})
        link1 = div.get("href")
        names1.append(name1.text)
        links1.append(link1)
            
            
    for div in soup.findAll('a',href=True, attrs={'class':'a-size-base a-link-normal s-no-hover a-text-normal'}):
        price1 = div.find('span', attrs={'class':'a-price-whole'})
        prices1.append(price1.text)
        
            
            
    q= min(prices1)
    p = prices1.index(q)
    comp.append(min(prices1))
    comp2.append(names1[p])
    comp3.append(links1[p])
    
    
    
amzp()

fp=[]

def flpc():
    driver.get(fin[1])

    content = driver.page_source
    soup = BeautifulSoup(content)
    for a in soup.findAll('a',href=True, attrs={'rel':'noopener noreferrer'}):
        link2 = a.get("href")
        links2.append(link2)

        try:
            name2 = a.find('div', attrs={'class':'_3wU53n'})
            names2.append(name2.text)
            
        except:
            name3 = a.find('a', attrs={'class':'_2cLu-l'})
            names2.append(a.text)
        try:    
            price3 = a.find('div', attrs={'class':'_1vC4OE'})
            prices2.append(int(price3.text.replace(',','')[1:]))
        
        except:
            price2 = a.find('div', attrs={'class':'_1vC4OE _2rQ-NK'})
            prices2.append(int(price2.text.replace(',','')[1:]))
            
            
                     
    mp2 =min(prices2)                     
    comp.append(mp2)
    comp2.append(names2[prices2.index(mp2)])
    comp3.append(links2[prices2.index(mp2)])

flpc()



print("\nMinimum on Amazon with price ₹",comp[0])
print(comp2[0],"\n")
azl = 'https://www.amazon.in/' + comp3[0]
print(azl)

print("\nMinimum on Flipkart with price ₹",comp[1])
print(comp2[1],"\n")
fpl = 'https://www.flipkart.com/' + comp3[1]
print(fpl)

driver.close()

df = pd.DataFrame({'Prices':comp,'names':comp2}) 
df.to_csv('prices.csv', index=False, encoding='utf-8')




message_text= 'Links for the cheapest options are\nAMAZON:\n' + azl+ '\nFLIPKART:\n' + fpl   #message you want to send
no_of_message=1 # no. of time you want the message to be send
num= input("Enter the number where you want to recieve the links\n")
fnum = '91'+num
moblie_no_list= [fnum] # list of phone number can be of any length

def element_presence(by,xpath,time):
    element_present = EC.presence_of_element_located((By.XPATH, xpath))
    WebDriverWait(driver, time).until(element_present)

def is_connected():
    try:
        # connect to the host -- tells us if the host is actually
        # reachable
        socket.create_connection(("www.google.com", 80))
        return True
    except :
        is_connected()
driver = webdriver.Chrome("/Users/Anil/chromedriver")
driver.get("http://web.whatsapp.com")
sleep(10) #wait time to scan the code in second

def send_whatsapp_msg(phone_no,text):
    driver.get("https://web.whatsapp.com/send?phone={}&source=&data=#".format(phone_no))
    try:
        driver.switch_to_alert().accept()
    except Exception as e:
        pass

    try:
        element_presence(By.XPATH,'//*[@id="main"]/footer/div[1]/div[2]/div/div[2]',30)
        txt_box=driver.find_element(By.XPATH , '//*[@id="main"]/footer/div[1]/div[2]/div/div[2]')
        global no_of_message
        for x in range(no_of_message):
            txt_box.send_keys(text)
            txt_box.send_keys("\n")

    except Exception as e:
        print("invailid phone no :"+str(phone_no))
for moblie_no in moblie_no_list:
    try:
        send_whatsapp_msg(moblie_no,message_text)

    except Exception as e:
        sleep(10)
        is_connected()
```
